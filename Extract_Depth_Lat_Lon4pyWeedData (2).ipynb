```python
from obspy import UTCDateTime
from obspy.clients.fdsn import Client
```

```python
client = Client("IRIS")
```

```python
startTime = UTCDateTime("2018-01-01T00:00:00")
endTime = UTCDateTime("2020-01-31T00:00:00")
cat = client.get_events(starttime=startTime, endtime=endTime, minmagnitude=2.5, maxmagnitude=10, minlongitude=63.2, maxlongitude=79.96,minlatitude=24.45, maxlatitude=44) #include filename='lat_lon_depth.txt' to save to raw file
```

```python
print(cat.__str__(print_all=True))
```
